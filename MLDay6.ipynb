```python
# task1: K-Means Clustering and PCA

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris

# =====================================================
# PART 1: K-Means on Mall Customers Dataset
# =====================================================

# Load dataset
mall = pd.read_csv("Mall_Customers.csv")

# Select Annual Income and Spending Score
X = mall[['Annual Income (k$)', 'Spending Score (1-100)']]

# -------------------------
# Elbow Method
# -------------------------
wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42, n_init=10)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(6,4))
plt.plot(range(1,11), wcss, marker='o')
plt.title("Elbow Method")
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.grid(True)
plt.show()

# -------------------------
# Apply K-Means (k=5)
# -------------------------
kmeans = KMeans(n_clusters=5, random_state=42, n_init=10)
labels = kmeans.fit_predict(X)
```

```python
#task-2

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans

# Load dataset
df = pd.read_csv("Mall_Customers.csv")

# Select features
X = df[['Annual Income (k$)', 'Spending Score (1-100)']]

# Apply K-Means
kmeans = KMeans(n_clusters=5, random_state=42, n_init=10)
df['Cluster'] = kmeans.fit_predict(X)

# Scatter plot
plt.figure(figsize=(8,6))

sns.scatterplot(
    data=df,
    x='Annual Income (k$)',
    y='Spending Score (1-100)',
    hue='Cluster',          # Colour-coded cluster labels
    palette='Set1',
    s=80
)

# Plot cluster centroids
plt.scatter(
    kmeans.cluster_centers_[:, 0],
    kmeans.cluster_centers_[:, 1],
    color='black',
    marker='X',
    s=200,
    label='Centroids'
)

plt.title("K-Means Customer Clusters")
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")
plt.legend(title="Cluster")
plt.grid(True)
plt.show()
```

```python
#task=4

from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
import numpy as np

# Load Iris dataset
iris = load_iris()
X = iris.data

# Apply PCA (reduce to 2 components)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Reconstruct the original data from the 2 principal components
X_reconstructed = pca.inverse_transform(X_pca)

# Print shapes
print("Original Data Shape      :", X.shape)
print("Reduced Data Shape       :", X_pca.shape)
print("Reconstructed Data Shape :", X_reconstructed.shape)

# Variance explained by each principal component
variance = pca.explained_variance_ratio_

print("\nVariance Explained by Each Component:")
for i, v in enumerate(variance):
    print(f"PC{i+1}: {v:.4f} ({v*100:.2f}%)")

# Total variance explained
print("\nTotal Variance Explained:",
      round(np.sum(variance), 4),
      f"({np.sum(variance)*100:.2f}%)")
```

```python
#practice sheet-1 customer segmentation mini-project using K-means

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans

# Load dataset
df = pd.read_csv("Mall_Customers.csv")

# Display first 5 records
print(df.head())

# Select features
X = df[['Annual Income (k$)', 'Spending Score (1-100)']]

# -----------------------------
# Elbow Method
# -----------------------------
wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42, n_init=10)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(7,5))
plt.plot(range(1,11), wcss, marker='o')
plt.title("Elbow Method")
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.grid(True)
plt.show()

# -----------------------------
# K-Means Clustering
# -----------------------------
kmeans = KMeans(n_clusters=5, random_state=42, n_init=10)

df['Cluster'] = kmeans.fit_predict(X)

# Display clustered data
print(df.head())

# -----------------------------
# Cluster Visualization
# -----------------------------
plt.figure(figsize=(8,6))

sns.scatterplot(
    data=df,
    x='Annual Income (k$)',
    y='Spending Score (1-100)',
    hue='Cluster',
    palette='Set1',
    s=80
)

# Plot centroids
plt.scatter(
    kmeans.cluster_centers_[:,0],
    kmeans.cluster_centers_[:,1],
    color='black',
    marker='X',
    s=250,
    label='Centroids'
)

plt.title("Customer Segmentation using K-Means")
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#practice sheet-2 apply PCA before KNNand compare accuracy with/without PCA

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report

# Load Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Split dataset
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# Feature Scaling
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# -------------------------------
# KNN Without PCA
# -------------------------------
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)

y_pred = knn.predict(X_test)

acc_without_pca = accuracy_score(y_test, y_pred)

print("Accuracy without PCA:", round(acc_without_pca * 100, 2), "%")

# -------------------------------
# Apply PCA
# -------------------------------
pca = PCA(n_components=2)

X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)

# -------------------------------
# KNN With PCA
# -------------------------------
knn_pca = KNeighborsClassifier(n_neighbors=5)
knn_pca.fit(X_train_pca, y_train)

y_pred_pca = knn_pca.predict(X_test_pca)

acc_with_pca = accuracy_score(y_test, y_pred_pca)

print("Accuracy with PCA:", round(acc_with_pca * 100, 2), "%")

# -------------------------------
# Variance Explained
# -------------------------------
print("\nExplained Variance Ratio:")
print(pca.explained_variance_ratio_)

print("\nTotal Variance Explained:",
      round(sum(pca.explained_variance_ratio_) * 100, 2), "%")

# -------------------------------
# Classification Report
# -------------------------------
print("\nClassification Report (Without PCA)")
print(classification_report(y_test, y_pred))

print("\nClassification Report (With PCA)")
print(classification_report(y_test, y_pred_pca))
```
