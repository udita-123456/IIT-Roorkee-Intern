```python
def add_task(task):
    with open("todo.txt", "a") as file:
        file.write(task + "\n")
    print(f"Added: '{task}'")
def view_tasks():
    print("\nYour To-Do List:")
    try:
        with open("todo.txt", "r") as file:
            tasks = file.readlines()
            if not tasks:
                print("List is empty!")
            else:
                for index, task in enumerate(tasks, 1):
                    print(f"{index}. {task.strip()}")
    except FileNotFoundError:
        print("No to-do list found yet. Add a task first!")
add_task("Review linear regression math")
add_task("Finish Python file handling tasks")
view_tasks()
```

```python
def safe_divide(number1, number2):
    try:
        result = number1 / number2
        return f"The result is: {result}"
        
    except ZeroDivisionError:
        return "Error: Mathematics strictly forbids dividing by zero!"
        
    except TypeError:
        return "Error: Please provide numbers, not text!"
print(safe_divide(10, 2))      # Normal division
print(safe_divide(10, 0))      # Triggers ZeroDivisionError
print(safe_divide(10, "five")) # Triggers TypeError
```

```python
with open("log.txt", "w") as file:
    file.write("[INFO] System booted up safely.\n")
    file.write("[WARNING] Memory usage high during matrix inversion.\n")
    file.write("[INFO] User 'admin' logged in.\n")
    file.write("[ERROR] Division by zero in numerical method!\n")
    file.write("[ERROR] File missing in directory.\n")
    file.write("[INFO] Scheduled backup completed.\n")
def analyze_logs(filepath):
    tally = {"INFO": 0, "WARNING": 0, "ERROR": 0}   
    try:
        with open(filepath, "r") as file:
            for line in file:
                if "[INFO]" in line:
                    tally["INFO"] += 1
                elif "[WARNING]" in line:
                    tally["WARNING"] += 1
                elif "[ERROR]" in line:
                    tally["ERROR"] += 1
                    
        print("Log Analysis Report")
        for log_type, count in tally.items():
            print(f"{log_type}: {count} occurrences")
            
    except FileNotFoundError:
        print("Error: Could not locate the log file.")

analyze_logs("log.txt")
```

```python
import csv
raw_data = [
    ["Student Name", "Abstract Algebra", "Real Analysis", "Field Theory"],
    ["Alice", 85, 90, 88],
    ["Bob", 72, 65, 70],
    ["Charlie", 95, 92, 98]
]
with open("midterm_marks.csv", "w", newline='') as file:
    writer = csv.writer(file)
    writer.writerows(raw_data)
def process_grades(in_file, out_file):
    processed_data = []
    
    try:
        with open(in_file, "r") as file:
            reader = csv.reader(file)
            header = next(reader) 
            header.append("Average")
            processed_data.append(header)
            for row in reader:
                scores = [int(row[1]), int(row[2]), int(row[3])]
                # Calculate the average
                average = sum(scores) / len(scores)  
                # Append the rounded average to the end of the student's row
                row.append(round(average, 2))
                processed_data.append(row)
        with open(out_file, "w", newline='') as file:
            writer = csv.writer(file)
            writer.writerows(processed_data)   
        print(f"Success! Averages calculated and saved to: {out_file}")    
    except FileNotFoundError:
        print(f"Error: The input file '{in_file}' is missing.")
process_grades("midterm_marks.csv", "final_results.csv")
```

```python
with open("study_notes.txt", "w") as file:
    file.write("Jean Piaget is known for his theory of cognitive constructivism.\n")
    file.write("The Cranmer abacus is a mathematical tool modified for the visually impaired.\n")
    file.write("Jerome Bruner introduced the concept of the spiral curriculum.\n")
    file.write("Bloom's taxonomy classifies educational learning objectives.\n")
    file.write("Standardized Braille literacy is crucial for independent learning.\n")
def search_file(filename, keyword):
    print(f"\nSearching for '{keyword}'")
    match_found = False
    
    try:
        with open(filename, "r") as file:
            # enumerate(file, 1) automatically gives us a line counter starting at 1
            for line_number, line in enumerate(file, 1):
                
                # We use .lower() on both so the search is case-insensitive
                if keyword.lower() in line.lower():
                    # .strip() removes the invisible newline character for clean printing
                    print(f"Line {line_number}: {line.strip()}")
                    match_found = True
                    
        if not match_found:
            print("Keyword not found in the document.")
            
    except FileNotFoundError:
        print(f"Error: The file '{filename}' is missing.")
search_file("study_notes.txt", "Bruner")
search_file("study_notes.txt", "braille") # Notice this is lowercase, but it will still find it!
search_file("study_notes.txt", "Vygotsky")
```
